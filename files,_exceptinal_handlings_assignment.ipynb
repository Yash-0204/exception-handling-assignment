Q.1 What is the difference between interpreted and compiled languages?

    Ans. Compiled Languages
    Definition: The code is translated entirely into machine code (binary) before it’s run.
    Process:
    You write code (source code).
    A compiler translates it into an executable file.
    You run the executable.
    Examples: C, C++, Rust, Go
    Pros:
    Faster execution (since it's already machine code).
    Can catch more errors at compile-time.
    Cons:
    Slower to test and debug (you have to recompile every time).
    Platform-dependent unless specially handled.  
    Interpreted Languages
    Definition: The code is read and executed line-by-line by an interpreter at runtime.
    Process:
    You write the code.
    The interpreter runs it directly.
    Examples: Python, JavaScript, Ruby, PHP
    Pros:
    Easier to test and debug.
    More flexible and cross-platform.
    Cons:
    Slower execution (since it's interpreted on the fly).
    Errors may only show up at runtime.

Q.9 F What is the difference between multithreading and multiprocessing?

    Ans. Multithreading (One Process, Many Threads)
    Runs multiple threads in a single process.
    Threads share the same memory space.
    Good for I/O-bound tasks (like reading files, network requests, etc.).
    Limited by the GIL (Global Interpreter Lock) in CPython, so not ideal for CPU-heavy tasks.

    Example use case: Downloading multiple web pages at the same time.

    Multiprocessing (Multiple Processes)
    Runs multiple processes, each with its own Python interpreter and memory space.
    Good for CPU-bound tasks (like heavy calculations, data processing).Bypasses the GIL, so it can use multiple CPU cores.

    Example use case: Performing heavy data analysis on large datasets.
